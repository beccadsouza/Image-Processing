```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
import math
```

# **Pot Hole 1 (PH1)**

## PH1 - Input Image

```
pothole = cv2.imread('reb_input_1.jpg')
print("Image Dimensions {0}x{1}".format(pothole.shape[0],pothole.shape[1]))
cv2_imshow(cv2.resize(pothole,(240,320)))
```

## PH1 - RGB to Grayscale

```
p_gray = cv2.cvtColor(pothole, cv2.COLOR_BGR2GRAY)
cv2_imshow(cv2.resize(p_gray,(240,320)))
```

## PH1 - Median Filtering

```
p_filtered = cv2.medianBlur(p_gray,5)
cv2_imshow(cv2.resize(p_filtered,(240,320)))
```

## PH1 - Histogram Calculation

```
histr = cv2.calcHist([p_filtered],[0],None,[256],[0,256]) 
plt.plot(histr)
```

## PH1 - Binarization

```
ret,p_binary = cv2.threshold(p_filtered,150,255,0)
cv2_imshow(cv2.resize(p_binary,(240,320)))
```

## PH1 - Dilation

```
kernel = np.ones((5,5), np.uint8) 
p_dilation = cv2.erode(p_binary, kernel, iterations=1)
cv2_imshow(cv2.resize(p_dilation,(240,320)))
```

## PH1 - Contour Based Segmentation

```
temp, contours, hierarchy = cv2.findContours(255-p_dilation.copy(),cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
```

```
segmented_image = pothole.copy()
cv2.drawContours(segmented_image, contours, -1, (0,255,0), 3)
cv2_imshow(cv2.resize(segmented_image,(240,320)))
```

## PH1 - Feature Extraction: Fitting Rectangles & Ellipses on potholes

```
max_contour = 0
max_w = 0
for i in range(len(contours)):
    (x, y, w, h) = cv2.boundingRect(contours[i])
    if w > max_w:
        max_contour = i
        max_w = w
```

```
output = pothole.copy()
(x, y, w, h) = cv2.boundingRect(contours[max_contour])
print("Height = ",h)
print("Width = ",w)
print("Area = ", cv2.contourArea(contours[max_contour]))
cv2.rectangle(output, (x,y), (x+w, y+h), (0,255,0), 6)
cv2.putText(output, "Pothole", (x,y), cv2.FONT_HERSHEY_PLAIN, 6, (0, 255, 0), 6, cv2.LINE_AA) 
cv2_imshow(cv2.resize(output,(240,320)))
```

```
output = pothole.copy()
(x,y),(MA,ma),angle = cv2.fitEllipse(contours[max_contour])
(x, y, w, h) = cv2.boundingRect(contours[max_contour])
print("Area = ", cv2.contourArea(contours[max_contour]))
print("Perimeter = ", cv2.arcLength(contours[max_contour],True))
print("Orientation Angle = ", angle)
ellipse=cv2.fitEllipse(contours[max_contour])
cv2.ellipse(output,ellipse,(0,255,0),6)
cv2.putText(output, "Pothole", (x,y), cv2.FONT_HERSHEY_PLAIN, 6, (0, 255, 0), 6, cv2.LINE_AA) 
cv2_imshow(cv2.resize(output,(240,320)))
```
